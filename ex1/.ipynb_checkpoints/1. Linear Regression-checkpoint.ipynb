## Linear Regression with one variable

In this part of this exercise, we will implement linear regression with one variable to predict profits for a food truck. Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet. The chain already has trucks in various cities and you have data for profits and populations from the cities. You would like to use this data to help you select which city to expand to next.  

The file ```ex1data1.txt``` contains the dataset for our linear regression problem. The first column is the population of a city and the second column is the profit of a food truck in that city. A negative value for profit indicates a loss.

### Plotting the Data

Before starting on any task, it is often useful to understand the data by visualizing it. For this dataset, you can use a scatter plot to visualize the data, since it has only two properties to plot (profit and population). (Many other problems that you will encounter in real life are multi-dimensional and canâ€™t be plotted on a 2-d plot.)

First, the dataset is loaded from the data file into the variables $X$ and $y$.

````python
x,y = np.loadtxt("data/ex1data1.txt",delimiter=",",unpack=True)
m = len(y) # number of training examples
````

Next, we call the ```plotData``` function to create a scatter plot of the data.

````python
def plotData(x,y):
    '''Plots the data points and gives the figure axes labels of population and profit'''
    plt.scatter(x,y,marker='x',c='red',label="Training Data") # Plot the data
    plt.xlabel("Population of City in 10,000s") # Set the x-axis label
    plt.ylabel("Profit in $10,000s") # Set the y-axis label
    
print("Plotting Data...")
plotData(x,y)
````

### Gradient Descent

In this part, we will fit the linear regression parameters $\theta$ to our dataset using gradient descent.

#### Update Equations

The objective of linear regression is to minimize the cost function

<center>$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$</center>

where the hypothesis $h_\theta(x)$ is given by the linear model

<center>$h_\theta(x) = \theta^Tx = \theta_0 + \theta_1x_1$</center>

Recall that the parameters of our model are the $\theta_j$ values. These are the values we will adjust to minimize cost $J(\theta)$. One way to do this is to use the batch gradient descent algorithm. In batch gradient descent, each iteration performs the update

<center>$\theta_j := \theta_j - \alpha\frac{1}{m} \sum_{i=m}^{m} (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}$  (simulataneously update $\theta_j$ for all $j$)</center>

With each step of gradient descent, our parameters $\theta_j$ come closer to the optimal values that will achieve the lowest cost $J(\theta)$.

#### Implementation

In the following lines, we add another dimension to our data to accomodate for the $\theta_0$ intercept term. We also initialize the initial parameters to 0 and the learning rate ```alpha``` to 0.01.

````python
X = np.column_stack((np.ones((m,1)),x)) # Add a column of ones to x
y = np.reshape(y,(m,1))
theta = np.zeros((2,1)) # Initialize fitting parameters

# Some gradient descent settings
iterations = 1500
alpha = 0.01
````

#### Computing the cost $J(\theta)$

As we perform gradient descent to learn minimize the cost function $J(\theta)$, it is helpful to monitor the convergence by computing the cost. In this section, we will implement a function to calculation $J(\theta)$ so that we can check the convergence of our gradient descent implementation. 

````python
def computeCost(X, y, theta):
    '''Computes the cost of using theta as the parameter for linear regression to fit the data points in X and y'''
    m = len(y) # Number of training examples
    
    h = np.dot(X,theta)
    
    J = (np.dot((h-y).T,(h-y)))/(2*m)
    
    return np.asscalar(J)

# Compute and display initial cost
print computeCost(X,y,theta)
````

#### Gradient Descent

Next, we will implement gradient descent in the function ```gradientDescent```. Keep in mind that the cost $J(\theta)$ is parameterized by the vector $\theta$, not $X$ and $y$. That is, we minimize the value of $J(\theta)$, by changing the values of the vector $\theta$, not $X$ and $y$. 

A good way to verify that gradient descent is working correctly is to look at the value of $J(\theta)$ and check that it is decreasing with each step. 

````python
print "Running Gradient Descent..."

def gradientDescent(X,y,theta,aplha,num_iters):
    '''Performs gradient descent to learn theta. Updates theta by taking num_iters gradient steps with learning rate alpha'''
    m = len(y) # Number of training examples
    J_history = np.zeros((num_iters,1))
    
    for iter in range(num_iters):
        h = np.dot(X,theta)
        
        theta = theta - (alpha/m)*(np.dot(X.T,h-y))
        
        J_history[iter] = computeCost(X,y,theta)
    
    return (theta,J_history)

# Run gradient descent
theta, J_history = gradientDescent(X,y,theta,alpha,iterations)

# Print theta to screen
print "Theta found by gradient descent: "
print("%f %f" %(theta[0,], theta[1,]))
````

````python
# Plot the linear fit
plotData(x,y)
plt.hold(True)
plt.plot(x,np.dot(X,theta),label="Linear Regression")
plt.legend(loc="lower right")
````

````python
# Predict values for population sizes of 35,000 and 70,000

predict1 = np.dot([1,3.5],theta)

print("For population = 35,000, we predict a profit of %f"%(predict1*10000))

predict2 = np.dot([1,7],theta)

print("For population = 70,000, we predict a profit of %f"%(predict2*10000))
````

### Visualizing $J(\theta)$

To understand the cost function $J(\theta)$ better, we will now plot the cost over a 2D grid of $\theta_0$ and $\theta_1$ values.

````python
print "Visualizing J(theta_0,theta_1)..."

# Grid over which we will calculate J
theta0_vals = np.linspace(-10,10,100)
theta1_vals = np.linspace(-1,4,100)

# Initialize J_vals to a matrix of 0's
J_vals = np.zeros((len(theta0_vals),len(theta1_vals)))

# Fill out J_vals
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        t = np.array([theta0_vals[i], theta1_vals[j]]).reshape(theta.shape)
        J_vals[i,j] = computeCost(X,y,t)
        
ax = Axes3D(plt.gcf())        
ax.plot_surface(theta0_vals,theta1_vals,J_vals)
ax.set_xlabel("theta_0")
ax.set_ylabel("theta_1")
````

````python
# Contour plot
plt.figure()
plt.contour(theta0_vals,theta1_vals,J_vals)
plt.xlabel("theta_0")
plt.ylabel("theta_1")
plt.hold(True)
plt.plot(theta[0],theta[1],marker='x',c='red',ms=10,mew=2)
````

## Linear Regression with multiple variables

In this part, we will implement linear regression with multiple variables to predict the prices of houses. Suppose you are selling your house and you want to know what a good market price would be. One way of doing this would be to first collect information on recent houses sold and make a model of housing prices.

The file ```ex1data2.txt``` contains a training set of housing prices in Portland, Oregon. The first column is the size of the house (in square feet), the second column is the number of bedrooms, and the third column is the price of the house.


### Feature Normalization

We will start by loading and displaying some values from this dataset. By looking at the values, note that house sizes are about 1000 times the number of bedrooms. When features differ by orders of magnitude, first performing feature scaling can make gradient descent converge much more quickly. Our task here is to write a function ```featureNormalize``` to

* Subtract the mean value of each feature in the dataset.

* After subtracting the mean, additionally scale (divide) the feature values by their respective standard deviations.

````python
print "Loading data..."
data = np.loadtxt("data/ex1data2.txt",delimiter=",")
x = data[:,:2]
y = data[:,2]
m = len(y)

# Print out some data points
print "First 10 examples from the dataset: "
print "x = "
print x[0:9]
print "y = "
print y[0:9]
````

````python
def featureNormalize(X):
    '''Returns a normalized version of X where the mean value for each feature is 0 and the standard deviation is 1.
    This is often a good preprocessing step to do when working with learning algorithms.'''
    X_norm = X
    mu = np.zeros((1,X.shape[1]))
    sigma = np.zeros((1,X.shape[1]))
    
    mu = np.mean(X,axis=0)
    sigma = np.std(X,axis=0)
    
    X_norm = (X - mu)/sigma
    
    return (mu, sigma, X_norm)

# Scale features and set them to zero mean
print "Normalizing features..."

mu, sigma, X = featureNormalize(x)

# Add intercept term to X
X = np.column_stack((np.ones((m,1)),X))
y = np.reshape(y,(m,1))
````

### Gradient Descent

````python
print "Running gradient descent..."

alpha = 0.01
num_iters = 400

theta = np.zeros((3,1))
theta, J_history = gradientDescent(X,y,theta,alpha,num_iters)

plt.plot(range(len(J_history)),J_history)
plt.xlabel("Number of iterations")
plt.ylabel("Cost J")
````

````python
# Display gradient descent result
print("Theta computed from gradient descent: ")
print theta

normalized_X = ([1650,3] - mu)/ sigma
normalized_X = np.insert(normalized_X,0,1)

price = np.asscalar(np.dot(normalized_X,theta))

print "\nPredicted a price of a 1650 sq ft, 3 br house (using gradient descent): " + str(price)
````

### Normal Equations

In the lecture videos, we have learned that the closed-form solution to linear regression is:

<center>$\theta = (X^TX)^{-1}X^Ty$</center>

Using this formula does not require any feature scaling, and we will get an exact solution in one calculation.

````python
print "Solving with normal equations..."

def normalEqn(X,y):
    '''Computes the closed-form solution to linear regression using the normal equations.'''
    theta = np.zeros((X.shape[1],1))
    
    theta = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
    
    return theta

data = np.loadtxt("data/ex1data2.txt",delimiter=",")
X = data[:,:2]
y = data[:,2]
m = len(y)

X = np.column_stack((np.ones((m,1)),X))
y = np.reshape(y,(m,1))

theta = normalEqn(X,y)

print("Theta computed from normal equations: ")
print theta

price = np.dot([1,1650,3],theta)

print "\nPredicted a price of a 1650 sq ft, 3 br house (using gradient descent): " + str(price)
````
